```python
# 读取d_labitems.csv.gz文件
d_labitems = pd.read_csv(mimic_path+'hosp/d_labitems.csv.gz', compression='gzip')

# 将itemid和label对应起来
itemid_to_label = dict(zip(d_labitems['itemid'], d_labitems['label']))

# 修改透视表的列索引为label
lab_pivot.columns = [itemid_to_label.get(x, x) for x in lab_pivot.columns]
```
